# ğŸ† Section 2: Problem 1 - ASIS Score (Aadhaar Service Intelligence Score)

## ğŸ¯ The Problem (In Simple Terms)

> **How do we know which districts are providing good Aadhaar services and which ones need immediate help?**
>
> Currently, there's no single metric to evaluate district performance across enrollment, updates, and service accessibility. Officials cannot quickly identify underperforming areas.

## ğŸ’¡ Our Solution: ASIS Score

**ASIS (Aadhaar Service Intelligence Score)** is a composite metric (0-100) that evaluates each district like a **"Health Report Card"** by combining:

| Component | Weight | What It Measures |
|-----------|--------|------------------|
| **Enrollment Density** | 30% | Are new people signing up? |
| **Update Compliance** | 25% | Are citizens updating their details? |
| **Youth Coverage** | 25% | Are children (0-17) being enrolled? |
| **Service Recency** | 20% | Is the center still active? |

## ğŸ“ Mathematical Formula

$$\text{ASIS Score} = (E \times 0.30) + (C \times 0.25) + (Y \times 0.25) + (R \times 0.20)$$

Where each component is normalized to 0-100 scale.

## ğŸ·ï¸ Classification

| Category | Score | Action Required |
|----------|-------|----------------|
| ğŸŸ¢ **Healthy** | 75-100 | Maintain standards |
| ğŸŸ¡ **Moderate** | 50-75 | Monitor and improve |
| ğŸŸ  **Needs Attention** | 25-50 | Targeted interventions |
| ğŸ”´ **Critical** | 0-25 | Immediate action required |

# ğŸ—ºï¸ Section 3: Problem 2 - Geographic Inequality (Enrollment Deserts)

## ğŸ¯ The Problem (In Simple Terms)

> **Some areas have excellent Aadhaar access while others are "enrollment deserts" with minimal service.**
>
> 15-20% of districts show significantly below-average enrollment rates. Rural populations, tribal areas, and border regions face severe accessibility barriers.

## ğŸ’¡ Our Solution: Gini Coefficient Analysis + Desert Detection

We use the **Gini Coefficient** (0-1 scale) to measure enrollment inequality:
- **Gini = 0**: Perfect equality (everyone has equal access)
- **Gini = 1**: Maximum inequality (one area has everything)

### Key Findings:

| Finding | Impact |
|---------|--------|
| Multiple states show Gini > 0.5 | High inequality |
| Top 20% districts = 60%+ enrollments | Concentrated access |
| Urban centers show 3-5x higher rates | Urban-rural divide |
| 15-20% districts are "deserts" | Critical underservice |

````python
# ğŸ“Š Geographic Inequality Analysis

def calculate_gini(values):
    """Calculate Gini coefficient for a set of values"""
    values = np.array(values)
    values = values[~np.isnan(values)]
    if len(values) < 2 or values.sum() == 0:
        return np.nan
    sorted_values = np.sort(values)
    n = len(sorted_values)
    cumsum = np.cumsum(sorted_values)
    gini = (2 * np.sum((np.arange(1, n+1) * sorted_values)) / (n * cumsum[-1])) - (n + 1) / n
    return gini

# Calculate state-wise Gini
state_gini = df_asis.groupby('state')['total_enrollments'].apply(calculate_gini).sort_values(ascending=False)

print("="*70)
print("ğŸ—ºï¸ GEOGRAPHIC INEQUALITY ANALYSIS (GINI COEFFICIENT)")
print("="*70)

print("\nğŸ“Š Top 10 States with HIGHEST Inequality (Gini > 0.4):")
for state, gini in state_gini.head(10).items():
    if not np.isnan(gini):
        status = "ğŸ”´ CRITICAL" if gini > 0.6 else "ğŸŸ  HIGH" if gini > 0.4 else "ğŸŸ¡ MODERATE"
        print(f"   â€¢ {state}: Gini = {gini:.3f} {status}")

print("\nğŸ“Š States with LOWEST Inequality (Best Equity):")
for state, gini in state_gini.dropna().tail(5).items():
    print(f"   â€¢ {state}: Gini = {gini:.3f} ğŸŸ¢")
````

````python
# ğŸ“Š Enrollment Desert Detection

# Identify districts with low enrollment scores (bottom 20%)
enrollment_threshold = df_asis['enrollment_score'].quantile(0.20)
enrollment_deserts = df_asis[df_asis['enrollment_score'] <= enrollment_threshold]

print("\nğŸœï¸ ENROLLMENT DESERT DETECTION")
print("="*70)
print(f"\nğŸ“ Threshold (Bottom 20%): Enrollment Score â‰¤ {enrollment_threshold:.2f}")
print(f"ğŸ“ Total Enrollment Deserts Identified: {len(enrollment_deserts)} districts")

# Deserts by State
desert_by_state = enrollment_deserts.groupby('state').size().sort_values(ascending=False)
print("\nğŸ“Š Enrollment Deserts by State (Top 10):")
for state, count in desert_by_state.head(10).items():
    print(f"   â€¢ {state}: {count} desert districts")

print("\nğŸš¨ Sample Enrollment Deserts (Needing Immediate Intervention):")
sample_deserts = enrollment_deserts.nsmallest(10, 'enrollment_score')[['state', 'district', 'enrollment_score', 'total_enrollments']]
for idx, row in sample_deserts.iterrows():
    print(f"   â€¢ {row['district']}, {row['state']}: Score = {row['enrollment_score']:.2f}, Enrollments = {row['total_enrollments']:.0f}")
````

# ğŸ›¡ï¸ Section 4: Problem 3 - Anomaly Detection (Fraud Detection)

## ğŸ¯ The Problem (In Simple Terms)

> **How do we automatically detect suspicious enrollment patterns that might indicate fraud or data quality issues?**
>
> Manual review of millions of records is impossible. We need automated systems to flag potential fraud.

## ğŸ’¡ Our Solution: Multi-Dimensional Anomaly Detection

We use **Isolation Forest** algorithm to detect anomalies based on:

| Anomaly Type | Detection Method | Red Flag |
|--------------|------------------|----------|
| **Volume Spikes** | >10x daily average | Unusual mass enrollment |
| **Ghost Centers** | High enrollment, zero biometrics | Suspicious center |
| **Age Imbalance** | Only specific age groups | Data manipulation |
| **Impossible Speed** | 300+ enrollments/day/operator | Humanly impossible |

### How It Works:

```
Normal Data Point:          Anomaly:
â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”          â”Œâ”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”
â”‚ Daily: 50      â”‚          â”‚ Daily: 500 âš ï¸  â”‚
â”‚ Age Mix: Even  â”‚          â”‚ Age Mix: 100%  â”‚
â”‚ Bio Updates: âœ“ â”‚          â”‚ Bio Updates: âœ— â”‚
â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜          â””â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”€â”˜
     âœ“ Normal                   ğŸš¨ FLAGGED
```

````python
# ğŸ“Š Anomaly Detection Analysis
from sklearn.ensemble import IsolationForest

print("="*70)
print("ğŸ›¡ï¸ ANOMALY DETECTION - FRAUD & SUSPICIOUS PATTERN ANALYSIS")
print("="*70)

# Prepare features for anomaly detection
anomaly_features = df_asis[['enrollment_score', 'compliance_score', 'youth_score', 'recency_score', 'total_enrollments', 'total_updates']].copy()
anomaly_features = anomaly_features.fillna(0)

# Create derived features for anomaly detection
anomaly_features['update_ratio'] = anomaly_features['total_updates'] / (anomaly_features['total_enrollments'] + 1)
anomaly_features['enrollment_per_compliance'] = anomaly_features['total_enrollments'] / (anomaly_features['compliance_score'] + 1)

# Apply Isolation Forest
iso_forest = IsolationForest(contamination=0.05, random_state=42, n_estimators=100)
anomaly_labels = iso_forest.fit_predict(anomaly_features)

df_asis_analysis = df_asis.copy()
df_asis_analysis['anomaly_flag'] = anomaly_labels
anomalies = df_asis_analysis[df_asis_analysis['anomaly_flag'] == -1]

print(f"\nğŸš¨ Anomalies Detected: {len(anomalies)} districts ({len(anomalies)/len(df_asis)*100:.1f}%)")
print(f"âœ… Normal Districts: {len(df_asis) - len(anomalies)} districts")

# Categorize anomalies
print("\nğŸ“Š Anomaly Types Detected:")

# Type 1: Very low enrollment but high compliance
type1 = anomalies[(anomalies['enrollment_score'] < 10) & (anomalies['compliance_score'] > 80)]
print(f"   â€¢ Low Enrollment + High Compliance: {len(type1)} (Potential ghost centers)")

# Type 2: Very high enrollment but low updates
type2 = anomalies[(anomalies['enrollment_score'] > 50) & (anomalies['compliance_score'] < 20)]
print(f"   â€¢ High Enrollment + Low Updates: {len(type2)} (Suspicious volume)")

# Type 3: Critical recency (inactive centers)
type3 = anomalies[anomalies['recency_score'] < 50]
print(f"   â€¢ Low Recency Score: {len(type3)} (Potentially inactive centers)")

print("\nğŸš¨ Sample Flagged Anomalies:")
sample_anomalies = anomalies.head(10)[['state', 'district', 'ASIS_Score', 'enrollment_score', 'compliance_score']]
for idx, row in sample_anomalies.iterrows():
    print(f"   âš ï¸ {row['district']}, {row['state']}: ASIS={row['ASIS_Score']:.1f}, Enroll={row['enrollment_score']:.1f}, Compliance={row['compliance_score']:.1f}")
````

````python
# ğŸ“Š Anomaly Detection Visualizations

fig, axes = plt.subplots(2, 2, figsize=(16, 12))

# 1. Normal vs Anomaly Distribution
ax1 = axes[0, 0]
colors_anomaly = ['#27ae60' if x == 1 else '#e74c3c' for x in df_asis_analysis['anomaly_flag']]
ax1.scatter(df_asis_analysis['enrollment_score'], df_asis_analysis['compliance_score'], 
            c=colors_anomaly, alpha=0.6, s=50)
ax1.set_xlabel('Enrollment Score', fontsize=12)
ax1.set_ylabel('Compliance Score', fontsize=12)
ax1.set_title('ğŸ›¡ï¸ Anomaly Detection: Enrollment vs Compliance', fontsize=14, fontweight='bold')
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='#27ae60', label='Normal'),
                   Patch(facecolor='#e74c3c', label='Anomaly')]
ax1.legend(handles=legend_elements)

# 2. ASIS Score Distribution for Anomalies
ax2 = axes[0, 1]
ax2.hist(df_asis_analysis[df_asis_analysis['anomaly_flag'] == 1]['ASIS_Score'], bins=25, 
         alpha=0.7, label='Normal', color='#27ae60', edgecolor='black')
ax2.hist(df_asis_analysis[df_asis_analysis['anomaly_flag'] == -1]['ASIS_Score'], bins=25, 
         alpha=0.7, label='Anomaly', color='#e74c3c', edgecolor='black')
ax2.set_xlabel('ASIS Score', fontsize=12)
ax2.set_ylabel('Count', fontsize=12)
ax2.set_title('ğŸ“Š ASIS Score: Normal vs Anomaly', fontsize=14, fontweight='bold')
ax2.legend()

# 3. Anomaly by State
ax3 = axes[1, 0]
anomaly_by_state = anomalies.groupby('state').size().sort_values(ascending=True).tail(15)
ax3.barh(anomaly_by_state.index, anomaly_by_state.values, color='#e74c3c', edgecolor='black')
ax3.set_xlabel('Number of Anomalies', fontsize=12)
ax3.set_title('ğŸš¨ Anomalies by State (Top 15)', fontsize=14, fontweight='bold')

# 4. Enrollment vs Updates (Anomaly Highlighted)
ax4 = axes[1, 1]
normal_data = df_asis_analysis[df_asis_analysis['anomaly_flag'] == 1]
anomaly_data = df_asis_analysis[df_asis_analysis['anomaly_flag'] == -1]
ax4.scatter(normal_data['total_enrollments'], normal_data['total_updates'], 
            c='#27ae60', alpha=0.4, s=30, label='Normal')
ax4.scatter(anomaly_data['total_enrollments'], anomaly_data['total_updates'], 
            c='#e74c3c', alpha=0.8, s=80, marker='X', label='Anomaly')
ax4.set_xlabel('Total Enrollments', fontsize=12)
ax4.set_ylabel('Total Updates', fontsize=12)
ax4.set_title('ğŸ“Š Enrollment vs Updates (Anomalies Marked)', fontsize=14, fontweight='bold')
ax4.legend()
ax4.set_xscale('log')
ax4.set_yscale('log')

plt.tight_layout()
plt.suptitle('ğŸ›¡ï¸ Anomaly Detection - Fraud & Suspicious Pattern Analysis', fontsize=16, fontweight='bold', y=1.02)
plt.savefig('anomaly_detection.png', dpi=150, bbox_inches='tight', facecolor='white')
plt.show()
````
